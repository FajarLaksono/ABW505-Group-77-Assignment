# 1. Summary

This analysis explores the **Cleveland Heart Disease Dataset** to build predictive models for identifying patients with heart disease. The dataset contains 303 anonymous patient records with 13 clinical features and 1 target variable indicating disease presence.

## 1.1. Key Highlights

### 1.1.1. Dataset Overview
- **Size**: 303 samples with 14 columns (13 features + 1 target)
- **Quality**: No missing values, clean numerical data
- **Target Distribution**: Relatively balanced (54.5% disease, 45.5% no disease)
- **Features**: Demographics (age, sex), clinical measurements (blood pressure, cholesterol, heart rate), and diagnostic results (chest pain type, ECG, exercise tests)

### 1.1.2. Analysis Approach
The analysis employed comprehensive exploratory data analysis (EDA) including:
- **Distribution Analysis**: Histograms, boxplots, violin plots for all features
- **Statistical Testing**: T-tests for continuous features and Chi-square tests for categorical features revealed that chest pain type (cp), thalassemia (thal), number of major vessels (ca), exercise-induced angina (exang), and slope are the most significant predictors
- **Correlation Analysis**: Heatmaps and scatter plots to identify feature relationships
- **Feature Importance Prediction**: Combined statistical scoring identified top predictive features 
Chest Pain, Thalassemia, and Number of major vessels are considered as very influential and predicted to be the top predictors. 

### 1.1.3. Model Development
Two classification models were developed with hyperparameter tuning via GridSearchCV:

1. **K-Nearest Neighbors (KNN)** - Winner
   - Accuracy: 83.61%
   - Precision: 78.05%
   - Recall: 96.97% (excellent disease detection)
   - F1-Score: 86.49%
   - ROC-AUC: 93.18%

2. **Decision Tree Classifier**
   - Accuracy: 75.41%
   - Precision: 75.00%
   - Recall: 81.82%
   - F1-Score: 78.26%
   - ROC-AUC: 81.93%

### 1.1.3. Key Findings
- **KNN outperformed Decision Tree** across all metrics, with an 8.2% higher accuracy
- **High recall (96.97%)** indicates KNN is excellent at this dataset, detecting disease cases, minimizing false negatives
- **Clinical insights**: Patients with heart disease show higher maximum heart rates (158.47 vs 139.10 bpm), lower ST depression, and are slightly younger on average
- **Most important features**: Chest pain type, thalassemia, and number of major vessels consistently emerged as top predictors. Predicted (Chest Pain, Thalassemia, and Number of major vessels), Decision Tree (Chest Pain, ST Depression, and Thalassemia), and KKN (Thalassemia, Number of major vessels, and Slope of the peak exercise ST segment).

### 1.1.4. Practical Implications
- The KNN model demonstrates strong potential for clinical screening applications
- High recall is critical in medical contexts to avoid missing disease cases
- Decision Tree model offers better interpretability for clinical decision-making
- Feature importance analysis aligns with clinical knowledge, validating the model's reliability

### 1.1.5. Limitations
- Small dataset (303 samples) may limit generalizability
- Results based on single train-test split; cross-validation provides more robust estimates
- Further feature engineering and ensemble methods could improve performance

#### 5.2.4. Outlier Detection and Quantification

```python
continuous_features = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

# Visualize outliers
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for i, feature in enumerate(continuous_features):
    Q1 = heart_disease_data[feature].quantile(0.25)
    Q3 = heart_disease_data[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Create boxplot
    sns.boxplot(y=heart_disease_data[feature], ax=axes[i], color='lightblue')
    
    # Add horizontal lines for bounds
    axes[i].axhline(y=lower_bound, color='red', linestyle='--', linewidth=2, label=f'Lower Bound: {lower_bound:.2f}')
    axes[i].axhline(y=upper_bound, color='red', linestyle='--', linewidth=2, label=f'Upper Bound: {upper_bound:.2f}')
    
    # Highlight outliers
    outliers = heart_disease_data[(heart_disease_data[feature] < lower_bound) | 
                                   (heart_disease_data[feature] > upper_bound)]
    if len(outliers) > 0:
        axes[i].scatter([0] * len(outliers), outliers[feature], color='red', s=100, 
                       zorder=5, label=f'Outliers: {len(outliers)}', marker='x', linewidths=3)
    
    axes[i].set_title(f'{feature} - Outlier Detection', fontweight='bold')
    axes[i].set_ylabel(feature)
    axes[i].legend(loc='best', fontsize=8)
    axes[i].grid(True, alpha=0.3, axis='y')

# Remove the last empty subplot
fig.delaxes(axes[5])

plt.suptitle('Outlier Detection using IQR Method', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# Outlier Detection using IQR method

print("=" * 80)
print("OUTLIER DETECTION USING IQR METHOD")
print("=" * 80)
print("\nMethod: Values outside [Q1 - 1.5*IQR, Q3 + 1.5*IQR] are considered outliers\n")

outlier_summary = []

for feature in continuous_features:
    Q1 = heart_disease_data[feature].quantile(0.25)
    Q3 = heart_disease_data[feature].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify outliers
    outliers = heart_disease_data[(heart_disease_data[feature] < lower_bound) | 
                                   (heart_disease_data[feature] > upper_bound)]
    
    outlier_count = len(outliers)
    outlier_percentage = (outlier_count / len(heart_disease_data) * 100)
    
    outlier_summary.append({
        'Feature': feature,
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'Lower Bound': lower_bound,
        'Upper Bound': upper_bound,
        'Outlier Count': outlier_count,
        'Outlier %': f"{outlier_percentage:.2f}%"
    })
    
    if outlier_count > 0:
        print(f"\n{feature.upper()}:")
        print(f"  Range: [{lower_bound:.2f}, {upper_bound:.2f}]")
        print(f"  Outliers: {outlier_count} ({outlier_percentage:.2f}%)")
        print(f"  Outlier values: {sorted(outliers[feature].values)}")

# Create summary DataFrame
outlier_df = pd.DataFrame(outlier_summary)
print("\n" + "=" * 80)
print("OUTLIER SUMMARY TABLE")
print("=" * 80)
print(outlier_df.to_string(index=False))
```

#### 5.2.5. Skewness & Kurtosis
This analysis evaluates the shape of distributions to determine if data transformation is needed for ML algorithms.

##### 5.2.5.1. Calculation

```python
# Calculate Skewness and Kurtosis
continuous_features = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

print("=" * 80)
print("SKEWNESS & KURTOSIS ANALYSIS")
print("=" * 80)
print("\nSKEWNESS (measures asymmetry of distribution):")
print("   Skewness  0:      Symmetric (Normal distribution)")
print("   Skewness > 0:      Right-skewed (tail on right)")
print("   Skewness < 0:      Left-skewed (tail on left)")
print("   |Skewness| > 1:    Highly skewed (transformation recommended)")
print("\nKURTOSIS (measures tailedness of distribution):")
print("   Kurtosis  3:      Normal distribution (mesokurtic)")
print("   Kurtosis > 3:      Heavy tails (leptokurtic)")
print("   Kurtosis < 3:      Light tails (platykurtic)")
print("   |Kurtosis - 3| > 2: Significant deviation from normal")
print("=" * 80 + "\n")

skewness_kurtosis_results = []

for feature in continuous_features:
    skewness = heart_disease_data[feature].skew()
    kurtosis = heart_disease_data[feature].kurtosis()
    
    # Determine distribution shape
    if abs(skewness) < 0.5:
        skew_interpretation = "Fairly Symmetric"
    elif abs(skewness) < 1:
        skew_interpretation = "Moderately Skewed"
    else:
        skew_interpretation = "Highly Skewed"
    
    if abs(kurtosis - 3) < 1:
        kurt_interpretation = "Near Normal"
    elif kurtosis > 3:
        kurt_interpretation = "Heavy Tails"
    else:
        kurt_interpretation = "Light Tails"
    
    # Transformation recommendation
    if abs(skewness) > 1 or abs(kurtosis - 3) > 2:
        transformation = "RECOMMENDED"
    else:
        transformation = "Not Needed"
    
    skewness_kurtosis_results.append({
        'Feature': feature,
        'Skewness': f"{skewness:.4f}",
        'Skew Type': skew_interpretation,
        'Kurtosis': f"{kurtosis:.4f}",
        'Kurt Type': kurt_interpretation,
        'Transformation': transformation
    })
    
    print(f"{feature.upper()}:")
    print(f"  Skewness:        {skewness:.4f} ({skew_interpretation})")
    print(f"  Kurtosis:        {kurtosis:.4f} ({kurt_interpretation})")
    print(f"  Transformation:  {transformation}")
    print()

# Create summary DataFrame
skew_kurt_df = pd.DataFrame(skewness_kurtosis_results)
print("=" * 80)
print("SKEWNESS & KURTOSIS SUMMARY TABLE")
print("=" * 80)
print(skew_kurt_df.to_string(index=False))
```

#### 5.3.4. Statistical Significance Tests
This analysis tests whether features significantly differ between
patients with and without heart disease.

Significance level (alpha): 0.05
  - p-value < 0.05: Feature is statistically significant
  - p-value >= 0.05: Feature is NOT statistically significant

Statistical significance tests are formal procedures that use a probability value ($p$-value) to determine whether the observed differences between data groups are likely due to a real effect or simply the result of random chance.

##### 5.3.4.1. T-TEST
Null Hypothesis (H0): Means are equal between disease and no-disease groups
Alternative Hypothesis (H1): Means are different between groups

```python
# T-TEST for Continuous Features

continuous_features = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

t_test_results = []

for feature in continuous_features:
    # Split data by target
    no_disease = heart_disease_data[heart_disease_data['target'] == 0][feature]
    disease = heart_disease_data[heart_disease_data['target'] == 1][feature]
    
    # Perform independent t-test
    t_statistic, p_value = stats.ttest_ind(no_disease, disease)
    
    # Calculate means
    mean_no_disease = no_disease.mean()
    mean_disease = disease.mean()
    mean_diff = mean_disease - mean_no_disease
    
    # Determine significance
    significant = "YES" if p_value < 0.05 else "NO"
    
    t_test_results.append({
        'Feature': feature,
        'Mean (No Disease)': f"{mean_no_disease:.2f}",
        'Mean (Disease)': f"{mean_disease:.2f}",
        'Difference': f"{mean_diff:.2f}",
        't-statistic': f"{t_statistic:.4f}",
        'p-value': f"{p_value:.6f}",
        'Significant': significant
    })
    
    print(f"{feature.upper()}:")
    print(f"  Mean (No Disease): {mean_no_disease:.2f}")
    print(f"  Mean (Disease):    {mean_disease:.2f}")
    print(f"  Difference:        {mean_diff:.2f}")
    print(f"  t-statistic:       {t_statistic:.4f}")
    print(f"  p-value:           {p_value:.6f}")
    print(f"  Significant:       {significant}")
    print()

# Create summary DataFrame
t_test_df = pd.DataFrame(t_test_results)
print("=" * 80)
print("T-TEST SUMMARY TABLE")
print("=" * 80)
print(t_test_df.to_string(index=False))
```

##### 5.3.4.1 CHI-SQUARE TEST
Null Hypothesis (H0): Feature is independent of target (no association)
Alternative Hypothesis (H1): Feature is associated with target

```python
# CHI-SQUARE TEST for Categorical Features

categorical_features = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']

chi_square_results = []

for feature in categorical_features:
    # Create contingency table
    contingency_table = pd.crosstab(heart_disease_data[feature], heart_disease_data['target'])
    
    # Perform chi-square test
    chi2_statistic, p_value, dof, expected_freq = stats.chi2_contingency(contingency_table)
    
    # Determine significance
    significant = "YES" if p_value < 0.05 else "NO"
    
    chi_square_results.append({
        'Feature': feature,
        'Chi-square': f"{chi2_statistic:.4f}",
        'Degrees of Freedom': dof,
        'p-value': f"{p_value:.6f}",
        'Significant': significant
    })
    
    print(f"{feature.upper()}:")
    print(f"  Chi-square statistic: {chi2_statistic:.4f}")
    print(f"  Degrees of freedom:   {dof}")
    print(f"  p-value:              {p_value:.6f}")
    print(f"  Significant:          {significant}")
    print(f"  Contingency table:")
    print(contingency_table)
    print()

# Create summary DataFrame
chi_square_df = pd.DataFrame(chi_square_results)
print("=" * 80)
print("CHI-SQUARE TEST SUMMARY TABLE")
print("=" * 80)
print(chi_square_df.to_string(index=False))
```

-  Red bars (p >= 0.05): Feature does NOT significantly differ between groups
- Features with p < 0.05 are strong candidates for the ML models.

### 5.4. Potential Feature Importance Analysis

This section synthesizes findings from the statistical tests and correlation analysis to predict which features will be most important for machine learning models.

#### 5.4.1. Feature Importance Ranking Based on Statistical Tests

```python
# Create feature importance ranking based on statistical tests and correlations

# First, let's get the correlation with target for all features
# Note: corr variable was created earlier in section 5.3.2
target_corr = corr['target'].drop('target').abs()  # Get absolute correlation with target

# Process continuous features (from t-test results)
continuous_importance = []
for result in t_test_results:
    feature = result['Feature']
    p_value = float(result['p-value'])
    t_stat = float(result['t-statistic'])
    mean_diff = abs(float(result['Difference']))  # Effect size for continuous
    correlation = target_corr[feature]
    
    continuous_importance.append({
        'Feature': feature,
        'Type': 'Continuous',
        'P-value': p_value,
        'Effect Size': mean_diff,
        'Correlation': correlation
    })

# Process categorical features (from chi-square results)
categorical_importance = []
for result in chi_square_results:
    feature = result['Feature']
    p_value = float(result['p-value'])
    chi2_stat = float(result['Chi-square'])  # Effect size for categorical
    correlation = target_corr[feature]
    
    categorical_importance.append({
        'Feature': feature,
        'Type': 'Categorical',
        'P-value': p_value,
        'Effect Size': chi2_stat,
        'Correlation': correlation
    })

# Combine all features
all_features = continuous_importance + categorical_importance
feature_importance_df = pd.DataFrame(all_features)

# Normalize metrics to 0-1 scale for combining
# For p-value: smaller is better, so we use (1 - p_value) but cap at significance level
feature_importance_df['P-value Score'] = feature_importance_df['P-value'].apply(
    lambda x: 1.0 if x < 0.001 else (1 - x) if x < 1 else 0
)

# Normalize effect size (min-max scaling)
effect_min = feature_importance_df['Effect Size'].min()
effect_max = feature_importance_df['Effect Size'].max()
feature_importance_df['Effect Size Score'] = (
    (feature_importance_df['Effect Size'] - effect_min) / (effect_max - effect_min)
)

# Correlation is already in 0-1 range (we took absolute value)
feature_importance_df['Correlation Score'] = feature_importance_df['Correlation']

# Calculate combined importance score (weighted average)
# Weights: P-value (30%), Effect Size (35%), Correlation (35%)
feature_importance_df['Combined Importance Score'] = (
    0.30 * feature_importance_df['P-value Score'] +
    0.35 * feature_importance_df['Effect Size Score'] +
    0.35 * feature_importance_df['Correlation Score']
)

# Sort by combined importance score
feature_importance_df = feature_importance_df.sort_values(
    'Combined Importance Score', ascending=False
).reset_index(drop=True)

# Display the ranking table
print("=" * 100)
print("FEATURE IMPORTANCE RANKING BASED ON STATISTICAL ANALYSIS")
print("=" * 100)
print()
print("Scoring Method:")
print("  - P-value Score: Higher for more significant features (p < 0.05)")
print("  - Effect Size Score: Normalized effect size (mean difference for continuous, chi-square for categorical)")
print("  - Correlation Score: Absolute correlation with target variable")
print("  - Combined Score: Weighted average (P-value: 30%, Effect Size: 35%, Correlation: 35%)")
print()
print("=" * 100)

# Create display dataframe with formatted values
display_df = feature_importance_df[[
    'Feature', 'Type', 'P-value', 'Effect Size', 'Correlation', 'Combined Importance Score'
]].copy()

display_df['P-value'] = display_df['P-value'].apply(lambda x: f"{x:.6f}")
display_df['Effect Size'] = display_df['Effect Size'].apply(lambda x: f"{x:.4f}")
display_df['Correlation'] = display_df['Correlation'].apply(lambda x: f"{x:.4f}")
display_df['Combined Importance Score'] = display_df['Combined Importance Score'].apply(lambda x: f"{x:.4f}")

print(display_df.to_string(index=True))
print("=" * 100)
```

     - May create biased trees if classes are imbalanced (not an issue here)
   - **Suitability for this dataset**:  **SUITABLE**
     - Interpretability is valuable for medical applications (doctors can understand decision logic)
     - Handles outliers well (we have some outliers in chol, oldpeak)
     - Works well with our mixed feature types
     - Can be regularized (max_depth, min_samples_split) to prevent overfitting

3. **SVM (Support Vector Machine)**
   - **Type**: Margin-based classifier
   - **Strengths**: 
     - Effective in high-dimensional spaces
     - Memory efficient
     - Versatile (different kernels for non-linear patterns)
   - **Weaknesses**: 
     - Requires feature scaling
     - Sensitive to outliers (affects support vectors)
     - Less interpretable (especially with non-linear kernels)
     - Requires careful hyperparameter tuning (C, gamma, kernel)
     - Computationally expensive for training
   - **Suitability for this dataset**:  **MODERATE**
     - Can work but requires more tuning effort
     - Less interpretable than decision trees
     - Sensitivity to outliers is a concern

4. **Naive Bayes**
   - **Type**: Probabilistic classifier
   - **Strengths**: 
     - Fast training and prediction
     - Works well with small datasets
     - Provides probability estimates
     - Simple implementation
   - **Weaknesses**: 
     - **Strong independence assumption**: Assumes all features are independent
     - Sensitive to how features are distributed
     - Performance degrades if independence assumption is violated
   - **Suitability for this dataset**:  **NOT SUITABLE**
     - **Major issue**: Our correlation matrix shows multiple correlated features (e.g., age & thalach: -0.40, thalach & slope, oldpeak & exang)
     - The independence assumption is clearly violated
     - Would likely underperform compared to other algorithms

**===> Selected Algorithms: KNN and Decision Tree**

**Based on the analysis above, we select:**

1. **KNN (K-Nearest Neighbors)**
2. **Decision Tree**

**===> Justification for Selection**

**Why KNN?**
- **Flexible decision boundaries**: Can capture complex, non-linear relationships between cardiac features and heart disease
- **No distributional assumptions**: Works well despite skewed features (chol, oldpeak)
- **Small dataset advantage**: With 303 samples, KNN is computationally manageable and can learn from local patterns
- **Non-parametric approach**: Doesn't assume any specific relationship between features and target
- **Proven for medical data**: KNN has been successfully applied in many medical classification tasks

**Why Decision Tree?**
- **Clinical interpretability**: Critical for medical applications - doctors can understand and validate the decision rules (e.g., "If chest pain type > 2 AND max heart rate > 150 THEN...")
- **Robust to data characteristics**: Handles outliers and skewed distributions naturally (our dataset has both)
- **No preprocessing required**: Works directly with our numeric features without scaling
- **Handles feature interactions**: Can capture complex interactions between features (e.g., age  max heart rate)
- **Complementary to KNN**: Uses a completely different algorithmic approach, allowing for meaningful comparison

**Why NOT SVM?**
- Less interpretable (black box model)
- Sensitive to outliers (we have outliers in multiple features)
- Requires extensive hyperparameter tuning
- Provides no additional algorithmic diversity compared to KNN (both use distance-like concepts)

**Why NOT Naive Bayes?**
- **Strong independence assumption is violated**: Our features are clearly correlated (correlation matrix shows multiple significant correlations)
- Would provide inferior results compared to KNN and Decision Tree
- Doesn't offer any unique advantages for this dataset

**===>  Expected Comparison**

By comparing KNN vs Decision Tree, we can analyze:
- **Instance-based vs Rule-based** learning
- **Distance-based vs Tree-based** decision making
- **Scalability trade-offs**: KNN (slow prediction) vs Decision Tree (fast prediction)
- **Interpretability trade-offs**: KNN (black box) vs Decision Tree (transparent rules)
- **Preprocessing requirements**: KNN (needs scaling) vs Decision Tree (no scaling needed)

This comparison will provide insights into which approach is more suitable for heart disease prediction and help understand the strengths and weaknesses of each paradigm.

#### 6.2.2. Train-Test Split

```python
# Split data into training and testing sets
# Using 80/20 split with stratification to maintain class balance
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,  # 20% for testing, 80% for training
    random_state=42,  # For reproducibility
    stratify=y  # Maintain class distribution in both sets
)

print("=" * 80)
print("TRAIN-TEST SPLIT")
print("=" * 80)
print(f"Training set size: {X_train.shape[0]} samples ({X_train.shape[0]/len(X)*100:.1f}%)")
print(f"Test set size: {X_test.shape[0]} samples ({X_test.shape[0]/len(X)*100:.1f}%)")
print(f"\nTraining set shape: {X_train.shape}")
print(f"Test set shape: {X_test.shape}")

print(f"\n{'Class Distribution':^80}")
print("=" * 80)
print(f"{'Training Set':<40} {'Test Set':<40}")
print("-" * 80)

train_dist = y_train.value_counts().sort_index()
test_dist = y_test.value_counts().sort_index()

for class_label in [0, 1]:
    train_count = train_dist[class_label]
    test_count = test_dist[class_label]
    train_pct = train_count / len(y_train) * 100
    test_pct = test_count / len(y_test) * 100
    
    print(f"Class {class_label}: {train_count} ({train_pct:.1f}%)".ljust(40) + 
          f"Class {class_label}: {test_count} ({test_pct:.1f}%)")
```

```python
# Initialize StandardScaler
scaler = StandardScaler()

# Fit scaler on training data ONLY, then transform both train and test
# IMPORTANT: Never fit on test data to avoid data leakage
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Convert back to DataFrame for easier inspection
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=X.columns, index=X_train.index)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=X.columns, index=X_test.index)

print("=" * 80)
print("FEATURE SCALING (StandardScaler)")
print("=" * 80)
print("Scaling formula: z = (x - mean) / std")
print("\nScaled training data shape:", X_train_scaled_df.shape)
print("Scaled test data shape:", X_test_scaled_df.shape)

print("\n" + "=" * 80)
print("BEFORE SCALING (Original Training Data)")
print("=" * 80)
print(X_train.describe().round(2))

print("\n" + "=" * 80)
print("AFTER SCALING (Scaled Training Data)")
print("=" * 80)
print("Note: Scaled data should have mean  0 and std  1")
print(X_train_scaled_df.describe().round(2))

print("\n" + "=" * 80)
print("DATA SUMMARY")
print("=" * 80)
print("For KNN: Use X_train_scaled and X_test_scaled")
print("For Decision Tree: Use X_train and X_test (no scaling needed)")
```

### 6.3. Model Training and Evaluation

#### 6.3.1. K-Nearest Neighbors (KNN) Classifier
KNN is an instance-based learning algorithm that classifies new data points based on the majority class of their k nearest neighbors in the feature space.

##### 6.3.1.1. Finding Optimal K Value

We'll use GridSearchCV to find the best value of k (number of neighbors) through cross-validation.

```python
# Define parameter grid for KNN
param_grid_knn = {
    'n_neighbors': list(range(1, 31)),  # Test k values from 1 to 30
    'weights': ['uniform', 'distance'],  # uniform: all neighbors equal weight, distance: closer neighbors have more weight
    'metric': ['euclidean', 'manhattan']  # Distance metrics
}

# Initialize KNN classifier
knn = KNeighborsClassifier()

# Perform Grid Search with 5-fold cross-validation
print("=" * 80)
print("KNN HYPERPARAMETER TUNING (GridSearchCV)")
print("=" * 80)
print("This may take a few moments...")
print(f"Testing {len(param_grid_knn['n_neighbors']) * len(param_grid_knn['weights']) * len(param_grid_knn['metric'])} combinations")

grid_search_knn = GridSearchCV(
    estimator=knn,
    param_grid=param_grid_knn,
    cv=5,  # 5-fold cross-validation
    scoring='accuracy',
    n_jobs=-1,  # Use all available CPU cores
    verbose=1
)

# Fit grid search on scaled training data
grid_search_knn.fit(X_train_scaled, y_train)

# Get best parameters
best_knn = grid_search_knn.best_estimator_
best_params_knn = grid_search_knn.best_params_
best_score_knn = grid_search_knn.best_score_

print("\n" + "=" * 80)
print("BEST PARAMETERS FOUND")
print("=" * 80)
print(f"Best k (n_neighbors): {best_params_knn['n_neighbors']}")
print(f"Best weights: {best_params_knn['weights']}")
print(f"Best metric: {best_params_knn['metric']}")
print(f"Best cross-validation accuracy: {best_score_knn:.4f} ({best_score_knn*100:.2f}%)")
```

##### 6.3.1.2. KNN Model Evaluation

```python
# Make predictions on test set
y_pred_knn = best_knn.predict(X_test_scaled)
y_pred_proba_knn = best_knn.predict_proba(X_test_scaled)[:, 1]

# Calculate metrics
accuracy_knn = accuracy_score(y_test, y_pred_knn)
precision_knn = precision_score(y_test, y_pred_knn)
recall_knn = recall_score(y_test, y_pred_knn)
f1_knn = f1_score(y_test, y_pred_knn)
roc_auc_knn = roc_auc_score(y_test, y_pred_proba_knn)
conf_matrix_knn = confusion_matrix(y_test, y_pred_knn)

print("=" * 80)
print("KNN MODEL PERFORMANCE ON TEST SET")
print("=" * 80)
print(f"Accuracy:  {accuracy_knn:.4f} ({accuracy_knn*100:.2f}%)")
print(f"Precision: {precision_knn:.4f} ({precision_knn*100:.2f}%)")
print(f"Recall:    {recall_knn:.4f} ({recall_knn*100:.2f}%)")
print(f"F1-Score:  {f1_knn:.4f} ({f1_knn*100:.2f}%)")
print(f"ROC-AUC:   {roc_auc_knn:.4f} ({roc_auc_knn*100:.2f}%)")

print("\n" + "=" * 80)
print("CONFUSION MATRIX")
print("=" * 80)
print(f"{'':>15} {'Predicted: No (0)':<20} {'Predicted: Yes (1)':<20}")
print(f"{'Actual: No (0)':<15} {conf_matrix_knn[0,0]:<20} {conf_matrix_knn[0,1]:<20}")
print(f"{'Actual: Yes (1)':<15} {conf_matrix_knn[1,0]:<20} {conf_matrix_knn[1,1]:<20}")

# True Negatives, False Positives, False Negatives, True Positives
tn_knn, fp_knn, fn_knn, tp_knn = conf_matrix_knn.ravel()

print("\n" + "=" * 80)
print("CONFUSION MATRIX BREAKDOWN")
print("=" * 80)
print(f"True Negatives (TN):  {tn_knn} - Correctly predicted No Disease")
print(f"False Positives (FP): {fp_knn} - Incorrectly predicted Disease (Type I Error)")
print(f"False Negatives (FN): {fn_knn} - Incorrectly predicted No Disease (Type II Error)")
print(f"True Positives (TP):  {tp_knn} - Correctly predicted Disease")

print("\n" + "=" * 80)
print("CLASSIFICATION REPORT")
print("=" * 80)
print(classification_report(y_test, y_pred_knn, target_names=['No Disease (0)', 'Disease (1)']))
```

#### 6.3.2. Decision Tree Classifier

Decision Tree is a rule-based algorithm that creates a tree-like model of decisions by recursively partitioning the feature space.

##### 6.3.2.1. Finding Optimal Hyperparameters

```python
# Define parameter grid for Decision Tree
param_grid_dt = {
    'max_depth': [3, 5, 7, 10, None],  # Maximum depth of tree (None = unlimited)
    'min_samples_split': [2, 5, 10, 20],  # Minimum samples required to split a node
    'min_samples_leaf': [1, 2, 4, 8],  # Minimum samples required at leaf node
    'criterion': ['gini', 'entropy']  # Splitting criterion (gini impurity or information gain)
}

# Initialize Decision Tree classifier
dt = DecisionTreeClassifier(random_state=42)

# Perform Grid Search with 5-fold cross-validation
print("=" * 80)
print("DECISION TREE HYPERPARAMETER TUNING (GridSearchCV)")
print("=" * 80)
print("This may take a few moments...")
print(f"Testing {len(param_grid_dt['max_depth']) * len(param_grid_dt['min_samples_split']) * len(param_grid_dt['min_samples_leaf']) * len(param_grid_dt['criterion'])} combinations")

grid_search_dt = GridSearchCV(
    estimator=dt,
    param_grid=param_grid_dt,
    cv=5,  # 5-fold cross-validation
    scoring='accuracy',
    n_jobs=-1,  # Use all available CPU cores
    verbose=1
)

# Fit grid search on original training data (Decision Tree doesn't need scaling)
grid_search_dt.fit(X_train, y_train)

# Get best parameters
best_dt = grid_search_dt.best_estimator_
best_params_dt = grid_search_dt.best_params_
best_score_dt = grid_search_dt.best_score_

print("\n" + "=" * 80)
print("BEST PARAMETERS FOUND")
print("=" * 80)
print(f"Best max_depth: {best_params_dt['max_depth']}")
print(f"Best min_samples_split: {best_params_dt['min_samples_split']}")
print(f"Best min_samples_leaf: {best_params_dt['min_samples_leaf']}")
print(f"Best criterion: {best_params_dt['criterion']}")
print(f"Best cross-validation accuracy: {best_score_dt:.4f} ({best_score_dt*100:.2f}%)")
```

##### 6.3.2.2. Decision Tree Model Evaluation

```python
# Make predictions on test set
y_pred_dt = best_dt.predict(X_test)
y_pred_proba_dt = best_dt.predict_proba(X_test)[:, 1]

# Calculate metrics
accuracy_dt = accuracy_score(y_test, y_pred_dt)
precision_dt = precision_score(y_test, y_pred_dt)
recall_dt = recall_score(y_test, y_pred_dt)
f1_dt = f1_score(y_test, y_pred_dt)
roc_auc_dt = roc_auc_score(y_test, y_pred_proba_dt)
conf_matrix_dt = confusion_matrix(y_test, y_pred_dt)

print("=" * 80)
print("DECISION TREE MODEL PERFORMANCE ON TEST SET")
print("=" * 80)
print(f"Accuracy:  {accuracy_dt:.4f} ({accuracy_dt*100:.2f}%)")
print(f"Precision: {precision_dt:.4f} ({precision_dt*100:.2f}%)")
print(f"Recall:    {recall_dt:.4f} ({recall_dt*100:.2f}%)")
print(f"F1-Score:  {f1_dt:.4f} ({f1_dt*100:.2f}%)")
print(f"ROC-AUC:   {roc_auc_dt:.4f} ({roc_auc_dt*100:.2f}%)")

print("\n" + "=" * 80)
print("CONFUSION MATRIX")
print("=" * 80)
print(f"{'':>15} {'Predicted: No (0)':<20} {'Predicted: Yes (1)':<20}")
print(f"{'Actual: No (0)':<15} {conf_matrix_dt[0,0]:<20} {conf_matrix_dt[0,1]:<20}")
print(f"{'Actual: Yes (1)':<15} {conf_matrix_dt[1,0]:<20} {conf_matrix_dt[1,1]:<20}")

# True Negatives, False Positives, False Negatives, True Positives
tn_dt, fp_dt, fn_dt, tp_dt = conf_matrix_dt.ravel()

print("\n" + "=" * 80)
print("CONFUSION MATRIX BREAKDOWN")
print("=" * 80)
print(f"True Negatives (TN):  {tn_dt} - Correctly predicted No Disease")
print(f"False Positives (FP): {fp_dt} - Incorrectly predicted Disease (Type I Error)")
print(f"False Negatives (FN): {fn_dt} - Incorrectly predicted No Disease (Type II Error)")
print(f"True Positives (TP):  {tp_dt} - Correctly predicted Disease")

print("\n" + "=" * 80)
print("CLASSIFICATION REPORT")
print("=" * 80)
print(classification_report(y_test, y_pred_dt, target_names=['No Disease (0)', 'Disease (1)']))
```

## 7. Conclusions and Recommendations

### 7.1. Summary of Findings

This analysis compared two machine learning algorithms (KNN and Decision Tree) for predicting heart disease:

**Key Results:**
- Both models achieved strong performance on the heart disease dataset
- The models demonstrate that the dataset contains meaningful patterns for prediction
- Multiple features contribute to accurate heart disease prediction
- Feature importance analysis conducted for both models reveals which diagnostic measurements are most predictive

**Model Comparison:**
- **K-Nearest Neighbors (KNN)**:
  - Instance-based learning approach
  - Requires feature scaling
  - Flexible decision boundaries
  - Feature importance assessed using permutation importance technique
  
- **Decision Tree**:
  - Rule-based learning approach
  - No feature scaling required
  - Highly interpretable (can visualize decision rules)
  - Robust to outliers and skewed distributions
  - Built-in feature importance from tree structure

### 7.2. Practical Implications
**For Medical Applications:**
- Decision Tree may be preferred in clinical settings due to its interpretability
- Doctors can understand and validate the decision rules
- Feature importance analysis (both permutation importance for KNN and built-in importance for Decision Tree) reveals which diagnostic measurements are most predictive, enabling clinicians to focus on the most critical indicators

**For Production Systems:**
- Consider ensemble methods (Random Forest, Gradient Boosting) to improve performance
- Implement cross-validation to ensure robust performance estimates
- Monitor model performance over time as new data arrives

### 7.3. Limitations

1. **Small Dataset**: 303 samples may not capture all variations in heart disease presentations
2. **Class Imbalance**: While relatively balanced (54.5% vs 45.5%), slight imbalance exists
3. **Feature Selection**: All features used; feature selection could potentially improve performance
4. **Hyperparameter Space**: More extensive hyperparameter tuning could yield better results
