```python
import sys
import ipynbname

# Get the full path of the current notebook
notebook_path = ipynbname.path()
print("Current path: ", notebook_path.parent)
sys.path.insert(0, notebook_path.parent.as_posix())

import book

while True:
    print("\n===== Book Management Menu =====")
    print("1. Add Book")
    print("2. List Books")
    print("3. Search Books")
    print("4. Save & Return to Main (or Exit)")
    
    choice = input("Select option (1-4): ").strip()
    
    if choice == "1":
        book.add_book()
    elif choice == "2":
        book.list_books()
    elif choice == "3":
        book.search_books()
    elif choice == "4":
        book.save_books()
        print("Returning to main program (or exiting from this demo)...")
        break
    else:
        print("Invalid option. Please select 1-4.")
```
