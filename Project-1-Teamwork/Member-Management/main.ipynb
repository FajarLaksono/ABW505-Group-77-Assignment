# Member Management System

```python
import sys
import ipynbname

# Get the full path of the current notebook
notebook_path = ipynbname.path()
print("Current path: ", notebook_path.parent)
sys.path.insert(0, notebook_path.parent.as_posix())

import member

members = member.load_members()

def main_menu():
    while True:
        print("\n===== MEMBER MANAGEMENT MENU =====")
        print("1. Register Member")
        print("2. List Members")
        print("3. Search Members")
        print("4. Does a member exist?")
        print("5. Save & Exit")

        choice = input("Choose an option: ").strip()
        
        # -------------------------
        # 1. Register Member
        # -------------------------
        if choice == "1":
            print("\n--- Register Member ---")
            member_id = input("Enter Member ID: ").strip()
            name = input("Enter Name: ").strip()
            email = input("Enter Email: ").strip()

            member.register_member(members, member_id, name, email)

        # -------------------------
        # 2. List Members
        # -------------------------
        elif choice == "2":
            member.list_members(members)

        # -------------------------
        # 3. Search Members
        # -------------------------
        elif choice == "3":
            keyword = input("Enter keyword to search: ").strip()
            results = member.search_members(members, keyword)

            if results:
                for m in results:
                    print(f"ID: {m['member_id']} | Name: {m['name']} | Email: {m['email']}")
            else:
                print("No matching members found.")

        # -------------------------
        # 4. Check if Member Exists
        # -------------------------
        elif choice == "4":
            member_id = input("Enter Member ID: ").strip()
            if member.does_member_id_exist(members, member_id):
                print("✅ Member exists.")
            else:
                print("❌ Member does not exist.")

        # -------------------------
        # 5. Save & Exit
        # -------------------------
        elif choice == "5":
            member.save_members(members)
            print("Data saved. Goodbye!")
            break

        else:
            print("Invalid option, please try again.")


# Start program
if __name__ == "__main__":
    main_menu()
```
