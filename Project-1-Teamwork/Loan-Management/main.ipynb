# Loan Management System

```python
import sys
import ipynbname

# Get the full path of the current notebook
notebook_path = ipynbname.path()
print("Current path: ", notebook_path.parent)
sys.path.insert(0, notebook_path.parent.as_posix())

import loan

# Load loans from CSV at startup
loan.load_loans()

while True:
    print("\n===== Loan Management Menu =====")
    print("1. Borrow Book")
    print("2. Return Book")
    print("3. List Loans")
    print("4. Save & Exit")
    
    choice = input("\nEnter your choice (1-4): ").strip()
    
    if choice == '1':
        loan.borrow_book()
    elif choice == '2':
        loan.return_book()
    elif choice == '3':
        loan.list_loans()
    elif choice == '4':
        loan.save_loans()
        print("\nData saved successfully. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 4.")
```
